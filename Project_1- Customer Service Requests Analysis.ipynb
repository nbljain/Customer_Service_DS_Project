## P-Value Test

```python
from scipy.stats import ttest_1samp
ttest,p_value=ttest_1samp(s2,15500)
print(p_value)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" We are rejecting null hypothesis :There is no difference between the mean of population and sample at 95 %")
else:
    print("We are accepting null hypothesis")
```
